```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
df = pd.read_csv('AbandonoEmpleados.csv',sep=';', index_col = 'id', na_values = '#N/D')
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
df.drop(columns = ['anos_en_puesto','conciliacion'], inplace = True)
```

```python
def graficos_eda_categoricos (cat):
    #number of rows
    from math import ceil 
    filas = ceil(cat.shape[1]/2)
    
    #define the grafic
    f,ax = plt.subplots(nrows =filas, ncols = 2, figsize = (16, filas*6))
    
    #integration of the plane in one dimension 
    ax = ax.flat
    
    # loop for the graphs ]
    for x, i in enumerate (cat):
        cat[i].value_counts().plot.barh(ax=ax[x])
        ax[x].set_title(i, fontsize = 12, fontweight = 'bold')
        ax[x].tick_params(labelsize=12)
```

```python
graficos_eda_categoricos(df.select_dtypes('O'))
```
